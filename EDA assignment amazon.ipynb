```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# load data

```python
df = pd.read_csv('amazon.csv')
```

```python
df.head()
```

```python
data = df.drop('product_id',axis = 1)
```

# Q.1 What is the average rating for each product category ?

```python
df['rating']=pd.to_numeric(df['rating'],errors = 'coerce')
```

```python
avg_rating = df.groupby('category')['rating'].mean().reset_index()
```

```python
avg_rating
```

```python
plt.figure(figsize=(10,80))
sns.barplot(df, x='rating', y='category')
```

# Q.2 What are the top rating_count products by category ?

```python
df_grouped = df.groupby(['category', 'product_name']).agg({'rating': 'mean','rating_count': 'sum'}).reset_index()
# Rename columns for clarity
df_grouped.columns = ['product_category', 'product_name', 'avg_rating', 'Total_Rating_Count']

# Sort by Product_Category, Total_Rating_Count in descending order, and Average_Rating in descending order
df_sorted = df_grouped.sort_values(by=['product_category', 'Total_Rating_Count', 'avg_rating'], ascending=[True, False, False])
```

```python
df_sorted 
```

# Q.3 What is the distribution of discounted prices vs. actual prices ?

```python
# Create histograms for actual and discounted prices
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].hist(df['actual_price'], bins=20, color='blue', alpha=0.7, label='actual Price')
ax[0].set_title('Distribution of Actual Prices')
ax[0].set_xlabel('Price')
ax[0].set_ylabel('Frequency')

ax[1].hist(df['discounted_price'], bins=20, color='red', alpha=0.7, label='Discounted Price')
ax[1].set_title('Distribution of Discounted Prices')
ax[1].set_xlabel('Price')
ax[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Q.4 How does the average discount percentage vary across categories ?

```python
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'],errors = 'coerce')
average_discounts = df.groupby('category')['discount_percentage'].mean().reset_index()
```

```python
average_discounts.head(10)
```

# Q.5 What are the most popular product names  ?

```python
# Calculate a popularity score
df['rating_count'] = pd.to_numeric(df['rating_count'],errors = 'coerce')
df['Popularity_Score'] = df['rating'] * df['rating_count']

# Sort products by popularity score in descending order
df_sorted = df.sort_values(by='Popularity_Score', ascending=False)

# Get top 10 products
df_top_10 = df_sorted.head(10)
```

```python
df_top_10
```

```python
# Plotting
plt.figure(figsize=(10,6))
plt.bar(df_top_10['product_name'], df_top_10['Popularity_Score'])
plt.xlabel('Product Name')
plt.ylabel('Popularity Score')
plt.title('Top 10 Most Popular Products')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# Q.8 What is the correlation between discounted_price and rating ?

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'],errors = 'coerce')
correlation_coefficient = df['discounted_price'].corr(df['rating'])
correlation_coefficient
```

# Q.9 What are the Top 5 categories based on the highest rating ?

```python
df_grouped = df.groupby('category')['rating'].mean().reset_index()

# Sort by average rating in descending order and get top 5 categories
df_sorted = df_grouped.sort_values(by='rating', ascending=False).head(5)
```

```python
df_sorted
```

```python
# Plotting
plt.figure(figsize=(10,6))
plt.bar(df_sorted['category'], df_sorted['rating'])
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Top Categories by Average Rating')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```
