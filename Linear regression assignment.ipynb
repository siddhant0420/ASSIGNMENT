# Q.1 What does R-squared represent in a regression model ?

```python
# > R-squared (R²) is a statistical measure in a regression model that shows the proportion of variance in the dependent variable
#   that can be explained by the independent variable.
# >R-squared values range from 0 to 1, often expressed as a percentage from 0% to 100%.
# --0% R-squared:-
#            The model does not explain any of the variation in the response variable around its mean. 
#            The mean of the dependent variable predicts the dependent variable as well as the regression model.
# --100% R-squared:-
#            The model explains all the variation in the response variable around its mean.
```

# Q.2 What are the assumptions of linear regression ?

```python
# > The assumptions of linear regression are conditions that should be met to draw inferences regarding the model estimates or prediction.
# ~~ Linear Relationship :- The core premise of multiple linear regression is the existence of a linear relationship between
#    the dependent variable and the independent variables134.
#    This can be viewed visually using scatterplots .
# ~~ Independence of errors : The errors of the response variables are uncorrelated with each other.
#    Correlation between sequential observations, or autocorrelation, can be an issue with time series data.
# ~~ No Multicollinearity:- It is essential to avoid high correlation between the independent variables, a condition known as multicollinearity.
#    This can be checked using correlation matrices, where correlation coefficients should ideally be below 0.80,
#    or the Variance Inflation Factor (VIF), with VIF values above 10 indicating problematic multicollinearity.
# ~~ Homoscedasticity :- The variance of error terms should be consistent across all levels of the independent variables.
#    A scatterplot of residuals versus predicted values should not display any discernible pattern, such as a cone-shaped distribution,
#    which would indicate heteroscedasticity. 
```

# Q .3 What is the difference between R-squared and Adjusted R-squared ?

```python
# R-squared: 
#           This measures the proportion of variance in the dependent variable explained by the independent variables in the model. 
#           A limitation of R-squared is that it always increases (or remains the same) as you add more independent variables to the model,
#           even if those variables dont actually improve the models predictive power

# Adjusted R-squared: This is a modified version of R-squared that adjusts for the number of predictors in the model.
#                     It only adds new predictors to the model if they improve the models predictive power.
#                     Adjusted R-squared increases when the new term improves the model more than would be expected by chance,
#                     and it decreases when a predictor improves the model by less than expected.
```

# Q .17 Why is MSE sensitive to outliers ?

```python
# . MSE (Mean Squared Error) is sensitive to outliers because it squares the errors (the difference between actual and predicted values).

# . Quaring the errors amplifies the effect of large deviations.This means that outliers, which have large errors, disproportionately increase the MSE.
# . A single data point with a significant difference between the actual and predicted value can heavily influence the MSE.
```

# Q.18 What is the role of homoscedasticity in linear regression ?

```python
# Homoscedasticity, meaning "same variance," is a central assumption in linear regression models.
#  It means that the error term (the noise or random disturbance in the relationship between the independent variables and the dependent variable)
#  is the same across all values of the independent variables.
#  Homoscedasticity is important for reliable linear regression analysis4. When this assumption is met, it's easier to estimate the standard deviation
#  and variance of a dataset.
```

# Q. 2 Write a Python script to calculate and print Mean Squared Error (MSE), Mean Absolute Error (MAE), and Root Mean Squared Error (RMSE) for a linear regression model. ?

```python
 import numpy as np
 from sklearn.linear_model import LinearRegression
 from sklearn.metrics import mean_squared_error, mean_absolute_error
 from sklearn.model_selection import train_test_split
 
 X = diamonds[['carat', 'depth', 'table', 'x', 'y', 'z']]  # Independent variable
 y = diamonds['price']    # Dependent variable
 

 # Split data into training and testing sets
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
 

 # Create a linear regression model
 model = LinearRegression()
 

 # Train the model
 model.fit(X_train, y_train)
 

 # Make predictions on the test set
 y_pred = model.predict(X_test)
 

 # Calculate MSE
 mse = mean_squared_error(y_test, y_pred)
 

 # Calculate MAE
 mae = mean_absolute_error(y_test, y_pred)
 

 # Calculate RMSE
 rmse = np.sqrt(mse)
 

 # Print the results
 print("Mean Squared Error (MSE):", mse)
 print("Mean Absolute Error (MAE):", mae)
 print("Root Mean Squared Error (RMSE):", rmse)
```

# Q.11  Write a Python script that fits polynomial regression models of different degrees to a synthetic dataset and compares their performance.

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(0)  
X = np.random.rand(100, 1)  
y = 3 + 2 * X + 0.5 * X**2 + np.random.randn(100, 1) / 1.5  

# Define degrees for polynomial regression
degrees = [1, 2, 3, 4]

# Initialize lists to store MSEs and models
mses = []
models = []

# Fit polynomial regression models of different degrees
for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)
    
    model = LinearRegression()
    model.fit(X_poly, y)
    
    y_pred = model.predict(X_poly)
    mse = mean_squared_error(y, y_pred)
    
    mses.append(mse)
    models.append(model)

# Print MSEs for each degree
for degree, mse in zip(degrees, mses):
    print(f"Degree {degree}: MSE = {mse}")

# Plot data and regression curves for each degree
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data', color='blue', alpha=0.5)

for degree, model in zip(degrees, models):
    poly_features = PolynomialFeatures(degree=degree)
    X_test = np.linspace(0, 1, 100).reshape(-1, 1)
    X_test_poly = poly_features.fit_transform(X_test)
    y_pred = model.predict(X_test_poly)
    
    plt.plot(X_test, y_pred, label=f'Degree {degree}')

plt.title('Polynomial Regression Models')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# Q.14  Write a Python script that uses the Variance Inflation Factor (VIF) to check for multicollinearity in a dataset with multiple features.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Function to calculate VIF
def calculate_vif(data):
    # Initialize VIF dictionary
    vif_dict = {}
    
    # Iterate over each feature in the data
    for i in range(data.shape[1]):
        # Extract the feature
        feature = data.iloc[:, i]
        
        # Create a design matrix with the feature and all other features
        design_matrix = data.drop(data.columns[i], axis=1)
        
        # Add a constant to the design matrix
        design_matrix = sm.add_constant(design_matrix)
        
        # Fit the linear model
        model = sm.OLS(feature, design_matrix).fit()
        
        # Calculate R-squared
        r_squared = model.rsquared
        
        # Calculate VIF
        vif = 1 / (1 - r_squared)
        
        # Store VIF in the dictionary
        vif_dict[data.columns[i]] = vif
    
    return vif_dict

# Example usage
if __name__ == "__main__":
    # Create a sample dataset
    np.random.seed(0)
    data = pd.DataFrame({
        'Feature1': np.random.rand(100),
        'Feature2': np.random.rand(100),
        'Feature3': np.random.rand(100),
        'Feature4': np.random.rand(100)
    })
    
    # Introduce some multicollinearity
    data['Feature4'] = data['Feature1'] + data['Feature2'] + np.random.rand(100) * 0.1
    
    # Calculate VIF
    vif_dict = calculate_vif(data)
    
    # Print VIF values
    print("Variance Inflation Factors:")
    for feature, vif in vif_dict.items():
        print(f"{feature}: {vif:.2f}")
        
    # Interpret VIF values
    print("\nInterpretation:")
    for feature, vif in vif_dict.items():
        if vif > 10:
            print(f"{feature} has high multicollinearity (VIF > 10).")
        elif vif > 5:
            print(f"{feature} has moderate multicollinearity (5 < VIF ≤ 10).")
        else:
            print(f"{feature} has low multicollinearity (VIF ≤ 5).")
```

# Q.18 Write a Python script that performs multiple linear regression on a synthetic dataset with 5 features. Print the R-squared score and model coefficients.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pandas as pd

# Generate synthetic dataset
np.random.seed(0)  # For reproducibility
n_samples = 1000
n_features = 5

# Generate features
X = np.random.rand(n_samples, n_features)

# Generate coefficients for the linear model
coefficients = np.random.rand(n_features)

# Generate target variable with some noise
y = np.dot(X, coefficients) + np.random.randn(n_samples) * 0.1

# Convert data to DataFrame for easier manipulation
df = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(1, n_features + 1)])
df['Target'] = y

# Define features and target
X = df.drop('Target', axis=1)
y = df['Target']

# Perform multiple linear regression
model = LinearRegression()
model.fit(X, y)

# Predict target values
y_pred = model.predict(X)

# Calculate R-squared score
r2 = r2_score(y, y_pred)

# Print results
print(f"R-squared Score: {r2}")
print("Model Coefficients:")
for feature, coefficient in zip(X.columns, model.coef_):
    print(f"{feature}: {coefficient}")
print(f"Intercept: {model.intercept_}")
```

# Q.24 Write a Python script that uses cross-validation to evaluate a Linear Regression model on a synthetic dataset.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score
import pandas as pd

# Generate synthetic dataset
np.random.seed(0)  # For reproducibility
n_samples = 1000
n_features = 5

# Generate features
X = np.random.rand(n_samples, n_features)

# Generate coefficients for the linear model
coefficients = np.random.rand(n_features)

# Generate target variable with some noise
y = np.dot(X, coefficients) + np.random.randn(n_samples) * 0.1

# Convert data to DataFrame for easier manipulation
df = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(1, n_features + 1)])
df['Target'] = y

# Define features and target
X = df.drop('Target', axis=1)
y = df['Target']

# Define the scoring metric (R-squared)
scorer = make_scorer(r2_score)

# Perform cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(LinearRegression(), X, y, cv=kf, scoring=scorer)

# Print results
print(f"Cross-Validation R-squared Scores Across Folds: {scores}")
print(f"Average R-squared Score: {np.mean(scores)}")
```
